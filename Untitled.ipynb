<h1>Zadanie 1</h1>
<p>Zadanie polegało na powtórzeniu eksperymentu z poprzedniej listy, w którym liczyliśmy iloczyn skalarny testując pare algorytmów</p>
<p>Teraz wprowadzamy niewielkie zmiany w danych i sprwadzamy jak wpływa to na wynik eksperymentu</p>

```julia
function sum_forward(type,x,y)
    sum = type(0.0)
    for i in 1:length(x)
        sum += x[i] * y[i]
    end 
    return sum
end

function sum_backward(type,x,y)
    sum = type(0.0)
    for i in length(x):-1:1
        sum += x[i] * y[i]
    end 
    return sum
end

function sum_desc(type,x,y)
    xy = x .* y
    xy_pos = sort(filter(i -> i > 0, xy), rev=true)
    xy_neg = sort(filter(i -> i < 0, xy))
    
    xy_pos_sum = type(0.0)
    for i in xy_pos
        xy_pos_sum += i
    end

    xy_neg_sum = type(0.0)
    for i in xy_neg
        xy_neg_sum += i
    end
                                            
    return xy_pos_sum + xy_neg_sum
end

function sum_asc(type,x,y)
    xy = x .* y
    
    xy_pos = sort(filter(i -> i > 0, xy))
    xy_neg = sort(filter(i -> i < 0, xy), rev=true)
                                          
    xy_pos_sum = type(0.0)
    for i in xy_pos
        xy_pos_sum += i
    end

    xy_neg_sum = type(0.0)
    for i in xy_neg
        xy_neg_sum += i
    end
                                                            
    return xy_pos_sum + xy_neg_sum                                                       
end

x2 = [2.718281828, −3.141592654, 1.414213562, 0.577215664, 0.301029995]
x = [2.718281828, −3.141592654, 1.414213562, 0.5772156649, 0.3010299957]
y = [1486.2497, 878366.9879, −22.37492, 4773714.647, 0.000185049]

for t in [Float32,Float64]
    a = [t(i) for i in x2]
    b = [t(i) for i in y]
    println(rpad("arytmetyka:",20),t)
    println(rpad("dokladna wartosc:",20),−1.00657107000000e-11)
    println(rpad("forward sum:",20),sum_forward(t,a,b))
    println(rpad("backward sum:",20),sum_backward(t,a,b))                                                                    
    println(rpad("desc sum:",20),sum_desc(t,a,b))
    println(rpad("asc sum:",20),sum_asc(t,a,b),"\n")
end

print("="^40)
print("\n\n Wyniki z poprzedniego eksperymentu\n\n")
                    
for t in [Float32,Float64]
    a = [t(i) for i in x]
    b = [t(i) for i in y]
    println(rpad("arytmetyka:",20),t)
    println(rpad("dokladna wartosc:",20),−1.00657107000000e-11)
    println(rpad("forward sum:",20),sum_forward(t,a,b))
    println(rpad("backward sum:",20),sum_backward(t,a,b))                                                                    
    println(rpad("desc sum:",20),sum_desc(t,a,b))
    println(rpad("asc sum:",20),sum_asc(t,a,b),"\n")
end
```

<h1>Zadanie 3</h1>
<p>W tym zadaniu rozwiązujemy równanie $Ax=b$ gdzie</p>
<p>$A$ jest macierzą Hilberta lub losową macierzą o stopniu $n$ z zadanym wskaźnikiem $c$</p>
<p>$x$ to wektor jedynek $(1...1)^T$</p>
<p>Następnie trzeba rozwiązac równanie za pomoca dwóch algorytmów</p>
<p>Eliminacja gausa: $x=A$ \ $b$, oraz $x=A^{-1}b$</p>

```julia
using LinearAlgebra


# Function generates the Hilbert matrix  A of size n,
#  A (i, j) = 1 / (i + j - 1)
# Inputs:
#	n: size of matrix A, n>=1
#
#
# Usage: hilb(10)
#
# Pawel Zielinski
function hilb(n::Int)
        if n < 1
            error("size n should be >= 1")
        end
        return [1 / (i + j - 1) for i in 1:n, j in 1:n]
end

println(rpad("n",10),rpad("b. met. gausa",30)rpad("b. met. inw.",30)rpad("rząd",10),rpad("wskaznik uwarunkowania",30))
println("="^50,"Macierz Hilberta","="^50)

for n in 2:20
    A = hilb(n)
    x = ones(n)
    b = A*x

    x_g = A\b
    x_i = inv(A)*b
    
    e_g = norm(x_g - x) / norm(x)
    e_i = norm(x_i - x) / norm(x)

    println(rpad(n,10),rpad(e_g,30),rpad(e_i,30),rpad(rank(A),10),rpad(cond(A),30))
end
```

```julia
using LinearAlgebra

# Function generates a random square matrix A of size n with
# a given condition number c.
# Inputs:
#	n: size of matrix A, n>1
#	c: condition of matrix A, c>= 1.0
#
# Usage: matcond(10, 100.0)
#
# Pawel Zielinski
function matcond(n::Int, c::Float64)
        if n < 2
         error("size n should be > 1")
        end
        if c< 1.0
         error("condition number  c of a matrix  should be >= 1.0")
        end
        (U,S,V)=svd(rand(n,n))
        return U*diagm(0 =>[LinRange(1.0,c,n);])*V'
end

println(rpad("n",10),rpad("b. met. gausa",30),rpad("b. met. inw.",30),rpad("rząd",10),rpad("wskaznik uwarunkowania",30))
println("="^50,"Macierz Losowa","="^50)
ns = [5,10,20]
cs = [1.0, 10.0, 10.0^3, 10.0^7, 10.0^12, 10.0^16]

for n in ns
    for c in cs
        A = matcond(n,c)
        x = ones(n)
        b = A*x

        x_g = A\b
        x_i = inv(A)*b
    
        e_g = norm(x_g - x) / norm(x)
        e_i = norm(x_i - x) / norm(x)

        println(rpad(n,10),rpad(e_g,30),rpad(e_i,30),rpad(rank(A),10),rpad(cond(A),30))
    end
end
```

<h2>Wnioski</h2>
<p>W przypadku macierzy Hilberta, wraz ze współczynnikem uwarunkowania rośnie także błąd przybliżenia. Zazwyczaj mniejszy błąd otrzymaliśmy metodą Gaussa</p>
<p>W przyadku losowej macierzy błąd przybliżenia także rośnie wraz z wskaźnikiem uwarunkowania, dodatkowo rośnie wraz ze wzrostem rozmiaru macierzy. Tempo wzorstu jest znaczenie mniejsze od eksperymentu dla macierzy Hilberta, rożnica między błedem w obliczeniach metodą
Guassa oraz metodą inwersji także jest mniejsza</p>

<h1>Zadanie5</h1>
<p>W tym zadaniu mamy podane równanie rekurencyjne $p_{n+1} = p_n + r*p_n*(1-p_n)$</p>
<p>Mamy sprawdzić ciągi w arytmetyce Float32,Float64 oraz ostatni sposób gdzie we Float32 wynik 10-tej iteracji obcinamy do 3 miejsc po przecinku</p>

```julia
f(p,r) = p + r*p*(1-p)
p = 0.01
r = 3

f32 = [Float32(x) for x in 0:40]
f32_c = [Float32(x) for x in 0:40]
f64 = [Float64(x) for x in 0:40]

f32[1] = Float32(p)
f32_c[1] = Float32(p)
f64[1] = Float64(p)

print(rpad("n",30),rpad("f32",30),rpad("f32 obciete",30),rpad("f64",40),"\n")
print("="^120,"\n")

for i in 2:41
    f32[i] = Float32(f(f32[i-1],Float32(r)))
    f32_c[i] = Float32(f(f32_c[i-1],Float32(r)))
    f64[i] = Float64(f(f64[i-1],Float64(r)))

    if i==11
        f32_c[i] = floor(f32_c[i],digits=3)
    end

    print(rpad(i-1,30),rpad(f32[i],30),rpad(f32_c[i],30),rpad(f64[i],30),"\n")
end


```

<h2>Wnioski</h2>
<p>Do iteracji około 15 wszystkie modele dają podobne wyniki, od tego momentu model z obcięciem znacząco odbiega od pozostałych</p>
<p>Od iteracji około 20 modele bez obcięcia zaczynają dać odbiegające od siebie wyniki</p>
<p>Te wszystkie różnice spowodowane są przez kumulujące się błędy w obliczeniach</p>
<p>Proces wyznaczania kolejnych wartości ciągu jest niestabilny ponieważ kumulujące się błedy w każdej iteracji powodują coraz większy błąd w kolejnej</p>
<p>Na taki stan rzeczy ma wpływ moment w którym popełnia się błąd, oraz jego rozmiar</p>

```julia
using Plots
f(c, x) = x^2 + c
d = [[-2,1],[-2,2],[-2,1.99999999999999],[-1,1],[-1,-1],[-1,0.75],[-1,0.25]]

xx = [x for x in 1:40]
y = [Float64(x) for x in 0:40]

for i in d
    y[1] = i[2]
    for x in 2:41
        y[x] = f(i[1], y[x-1])
    end
    p = plot(xx, y[2:41],seriestype=:scatter)
    display(p)
    sleep(1)
end
```
